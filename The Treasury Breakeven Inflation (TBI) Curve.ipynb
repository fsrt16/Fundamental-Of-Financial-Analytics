```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
```

```python
df = pd.read_csv('F.csv')
taget = 'target'
nrows , ncols = df.shape
df.head()
```

```python
year = list(df['YEAR'])
rate = list(df['AVG avg_interest_among_all_matuarity'])
plt.bar(year,rate)
plt.xlabel("YEAR")
plt.ylabel("AVG avg interest among all matuarity Domain")
plt.title('The Treasury Breakeven Inflation (TBI) Curve')
plt.show()
```

```python
plt.pie(rate, labels=year, 
    autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("The Treasury Breakeven Inflation (TBI) Curve")
plt.show()
```
