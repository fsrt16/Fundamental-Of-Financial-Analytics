```python
def Identify(dataset):
    data_type=dataset.dtypes
    t=dict(data_type)
    new_dataset=pd.DataFrame()
    for k,v in t.items():
        df=pd.DataFrame()
        df[k]=dataset[k]
        #Specially for Tathagat this will replace every random shit character in that column with NaN
        df=df.replace(regex='[?#@!*()+-]',value=np.nan)
        if v=='object' or v=='bool':
            imp_most = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
            df=pd.DataFrame(imp_most.fit_transform(df))
            df.columns=[k]
            new_dataset[k]=df[k]
        else:
            imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
            df=pd.DataFrame(imp_mean.fit_transform(df))
            df.columns=[k]
            new_dataset[k]=df[k]
    return new_dataset
def preprocess(df):
      global numeric_variables , categorical_variables
      numeric_variables = list(df.select_dtypes(include=['int64','float']).columns.values)
      df[numeric_variables]=df[numeric_variables].apply(lambda x: x.fillna(x.mean()),axis=0)
      categorical_variables = list(df.select_dtypes(exclude=['int64','float','bool']).columns.values)
      df[categorical_variables]=df[categorical_variables].apply(lambda x: x.fillna(x.mode()),axis=0)
      df = pd.get_dummies(df, prefix_sep="__",columns=categorical_variables)  
      return df
def Automate_Preprocessing(df):
    df = Identify(df)
    df = preprocess(df)
    global target
    target = df['Liquidity Weight Index']
    df=df.drop('Liquidity Weight Index',axis=1)
    df = preprocess(df)
    return df
```

```python
def normalizedplot(dr,target=target):
    sns.set(style="white", palette="muted", color_codes=True)
    rs = np.random.RandomState(10)
    f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
    sns.despine(left=True)
    color = ["b","r","g","m"]
    i=0
    sns.distplot(dr.iloc[:,i], color=color[i],ax=axes[0, 0],norm_hist=True)
    i=i+1
    sns.distplot(dr.iloc[:,i], color=color[i],ax=axes[0, 1],norm_hist=True)
    i=i+1
    sns.distplot(dr.iloc[:,i], color=color[i],ax=axes[1, 0],norm_hist=True)
    i=i+1
    sns.distplot(dr.iloc[:,i], color=color[i],ax=axes[1, 1],norm_hist=True)
    import matplotlib.pyplot as plt1
    plt1.savefig('normalizedplot.png')
    
       
def boxplot(dr,target=target):
    sns.set(style="white", palette="muted", color_codes=True)
    rs = np.random.RandomState(10)
    f, axes = plt.subplots(2, 2, figsize=(7, 7), sharex=True)
    sns.despine(left=True)
    color = ["b","r","g","m"]
    for i in range(0,len(dr)):
        if i==0:
            sns.boxplot(x=target , y=dr.iloc[:,i],ax=axes[0, 0])
        if i==1:
            sns.boxplot(x=target , y=dr.iloc[:,i],ax=axes[0, 1])
        if i==2:
            sns.boxplot(x=target , y=dr.iloc[:,i],ax=axes[1, 0])
        if i==3:
            sns.boxplot(x=target , y=dr.iloc[:,i],ax=axes[1, 1])
    import matplotlib.pyplot as plt1
    plt1.savefig('boxplot.png')
def countplot(df,target=target):
    global conp
    conp = df.shape[1]
    for i in range(0,df.shape[1]):
        ax = sns.countplot(x=df.iloc[:,i],hue=target)
        import matplotlib.pyplot as plt
        plt.show()
        fig = ax.get_figure()
        fig.savefig("output"+str(i)+".png")
    
```

```python
def feature_univariate_numeric_features(X,Y=target):
    from sklearn.feature_selection import SelectKBest
    from sklearn.feature_selection import f_classif
    test = SelectKBest(score_func=f_classif, k=4)
    fity = test.fit(X, Y)
    features = fity.transform(X)
    nm = (X.columns[fity.get_support()]) 
    global dr 
    dr= pd.DataFrame(features,columns=nm)
    normalizedplot(dr)
    boxplot(dr)

def feature_sel_plot_bivariate_target_to_numeric_features(features,Y):
    for ft in features :
        sns.violinplot(x=Y, y=features[ft])
        fig, ax= plt.subplots(figsize=(10, 8))  
        sns.stripplot(Y, features[ft], jitter=True, ax=ax)  
        ax.set_title('Strip plot')  
        plt.show()
        import matplotlib.pyplot as plt1
        plt1.savefig('bivariate.png')
   
```
